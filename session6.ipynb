```python
# creating a fuction (with docstring)
def is_even(num):
    """
    This fuction returns if a given number is odd or even
    input - any valid integer
    output - odd/even
    created on - 10th May 2024
    """
    if type(num) == int:
      if num % 2 == 0:
          return 'even'
      else: 
          return 'odd'
    else: 
        return 'wrong data type inputted'
```

```python
# function
# function_name(input)
for i in range(1,11):
    x = is_even(i)
    print(x)
    
```

```python
# **kwargs
def display(**kwargs):
    for (key,value) in kwargs.items():
        print(key,'->',value)
        
```

```python
# lambda function
# x -> x^2
a = lambda x:x**2
a(2)
```

```python
# x,y -> x+y
a = lambda x,y:x+y
a(5,2)
```

```python
# check if a string has 'a'
a = lambda s:'a' in s
a('hello')
```

```python
# odd or even
a = lambda x:'even' if x%2 == 0 else 'odd'
a(6)
```

```python
# higher order functions
def square(x):
    return x**2
def transform(f,L):
    output = []
    for i in L:
        output.append(f(i))
print(output)
L = [1,2,3,4,5]
transform(square,L)

```

```python
# map
# square the item of a lists
list(map(lambda x:x**2,[1,2,3,4,5]))
```

```python
# filter
# number greater than 5
L =[1,2,3,5,6,7]
list(filter(lambda x:x>5,L))
```
